Las aseguradoras usan una gran cantidad de variables diferentes para calcular el precio de la prima, ya que este ultimo tiene que ser mayor al nivel de participacin de la aseguradora para poder obtener utilidades.

**Prima Base**
Valor real del riesgo asegurado, calculado con las probabilidades de que se necesite un servicio medico mas un porcentaje del costo de un servicio medico calculado por medio de una formula estadistica.

**Gastos Adicionales**
Todos aquellos Gastos adicionales a los antes mencionados.

**Precio de la prima= Prima Base + Gastos Adicionales**

Por lo que la siguiente funcin cumple con el clculo de la participacin de la aseguradora en un Servicio Medico

```python
#Clculo de la Prima Base

def prima(E,G,C,V):
    if E>18 and E<=30:
        P=0.015
    else:
        if E>30 and E<=40:
            P=0.02    
        elif E>40 and E<=55:
            P=0.025 
        elif E>55 and E<=65:
            P=0.03
        else:
            P=0
    if G==1:
        P=P+0.03
    else:
        P=P+0.01
    if C==1:
        P=P+0.27
    else:
        P=P+0.15
        
    return((P*V)/(1/22))
        
```

```python
#SIMULACIN DE LA PRIMA USANDO MONTECARLO
N=1000
E=np.random.uniform(18,100,size=1000)
G=np.random.uniform(1,2,size=1000)
C=np.random.uniform(1,2,size=1000)
V=np.random.uniform(4000,10000,size=1000)
A =list(map(lambda i: prima(E[i],G[i],C[i],V[i]), np.arange(N)))
plt.hist(A,density=True)
plt.show()
```

```python
sns.distplot(A, color="b", fit=st.beta, norm_hist=True)
```

```python
#Prueba de distribucin prima
prueba1 = st.beta.fit(A)
plot1 = st.probplot(A,sparams=(prueba1[0],prueba1[1]), dist="beta",fit=True, plot=plt)
plt.grid()
plt.xlabel('Beta theorical quantiles')
plt.ylabel('Data theorical quantiles')
plt.show()
```

```python
#chi2 para data1
[freq,x,p]=plt.hist(A,bins=100,density=True)
plt.show()
# Se obvia el ltimo valor de x para obtener exactamente J muestras de x
x = x[:-1]
y = st.beta.pdf(x,a=prueba1[0],b=prueba1[1],loc =prueba1[2], scale =prueba1[3])
Ei=x*y
X2=st.chisquare(freq,Ei)#chisquared
print('Valor Chi2 = ',X2)
#obtener valor de chi2 para aceptar o rechazar Ho. si p-value<0.05 rechazamos Ho
```

## Se Acepta Ho ya que nuestro pvalue=0.5800320481085203>0.05
## Por lo tanto decidimos que lo modelaremos con distribucion Beta ya que se puede modelar con esa

```python
#FUNCIN PARA CALCULAR LOS INGRESOS GENERANDO 1,000 PRIMAS ALEATORIAS

def ingreso(x):
    N=10000
    E=np.random.uniform(18,100,size=N)
    G=np.random.uniform(1,2,size=N)
    C=np.random.uniform(1,2,size=N)
    V=np.random.uniform(4000,100000,size=N)
    A =list(map(lambda i: prima(E[i],G[i],C[i],V[i]), np.arange(N)))
    Ingreso = sum(A)
    return Ingreso

#SIMULACIN DE LOS INGRESOS
B = list(map(lambda i: ingreso(i), np.arange(N)))
plt.hist(B,density=True)
plt.show()
```
